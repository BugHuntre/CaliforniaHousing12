#### Pipeline for Feature Engineering

```python
from sklearn.preprocessing import FunctionTransformer
def column_ratio(X):
    return X[:,[0]]/X[:,[1]]
def ratio_name(function_transformer,feature_names_in):
    """To give the feature names out"""
    return ['ratio']
def ratio_pipeline():
    return make_pipeline(
        SimpleImputer(strategy='median'),
        FunctionTransformer(column_ratio,feature_names_out = ratio_name),
        StandardScaler()
    )
    
```

```python
from sklearn.preprocessing import FunctionTransformer
log_pipeline = make_pipeline(
    SimpleImputer(strategy = 'median'),
    FunctionTransformer(np.log,feature_names_out = "one-to-one"),
    StandardScaler()
)
default_num_pipeline = make_pipeline(
    SimpleImputer(strategy = 'median'),
    StandardScaler()
)
```

```python
log_pipeline
```

```python
from sklearn.compose import make_column_selector

preprocessor = ColumnTransformer([
    ( 'bedrooms', ratio_pipeline(), [df.columns.get_loc(col) for col in ['total_bedrooms', 'total_rooms']] ),
    ( 'rooms_per_house', ratio_pipeline(), [df.columns.get_loc(col) for col in ['total_rooms', 'households']] ),
    ( 'people_per_house', ratio_pipeline(), [df.columns.get_loc(col) for col in ['population', 'households']] ),
    ( 'log', log_pipeline, [df.columns.get_loc(col) for col in ["total_bedrooms", "total_rooms", "households", "population", "median_income"]] ),
    ( 'cat', cat_pipeline, make_column_selector(dtype_include=object) )
], remainder=default_num_pipeline)
```

```python
preprocessor
```

```python
df_prepared = preprocessor.fit_transform(df)
```

## Model Training
1. Linear Regression
2. KNeighbourRegressor
3. DecisionTreeREgressor
4. RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_absolute_error,mean_squared_error
```

```python
def evaluate_model(true,predicted):
    mae = mean_absolute_error(true,predicted)
    mse = mean_squared_error(true,predicted)
    rmse = np.sqrt(mean_squared_error(true,predicted))
    r2_square = r2_score(true,predicted)
    return mae,rmse,r2_square,mse
```

```python
models = {
    "Linear Regression": LinearRegression(),
    "DecisionTree": DecisionTreeRegressor(),
    "K-Neghbors Regressor": KNeighborsRegressor(),
    "RandomForest Regressor": RandomForestRegressor()
}
```

```python
models
```

```python
for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train,y_train)
    # prediction
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    # model_evaluation 
    train_mae, train_rmse, train_r2_score, train_mse = evaluate_model(y_train,y_train_pred)
    test_mae, test_rmse, test_r2_score, test_mse = evaluate_model(y_test,y_test_pred)

    print(list(models.keys())[i])
    # evaluation print
    print("****Model Performance****")
    print("1. Mean Absolute error: {:.3f}".format(train_mae))
    print("2. Root Mean Squared error: {:.3f}".format(train_rmse))
    print("3. r2_score: {:.3f}".format(train_r2_score))
    print("4. Mean Squared error: {:.3f}".format(train_mse))
    print("*********************************************")
    print("1. Mean Absolute error: {:.3f}".format(test_mae))
    print("2. Root Mean Squared error: {:.3f}".format(test_rmse))
    print("3. r2_score: {:.3f}".format(test_r2_score))
    print("4. Mean Squared error: {:.3f}".format(test_mse))
```
